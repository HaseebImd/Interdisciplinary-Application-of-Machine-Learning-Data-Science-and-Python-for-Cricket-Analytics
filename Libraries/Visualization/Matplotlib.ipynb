<h2><center>Markers</center></h2>

You can use the keyword argument <code>marker</code> to emphasize each point with a specified marker:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o')
plt.show()
```

<h2><center>Matplotlib Labels and Title</center></h2>

With Pyplot, you can use the <code>xlabel()</code> and <code>ylabel()</code> functions to set a label for the x- and y-axis.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([120, 150, 190, 102, 80, 29, 100, 210, 220, 210])

plt.plot(x, y)

#title of the Graph
plt.title("Total mathes of a Player and Maximum Runs in one Match")

#x-axis label
plt.xlabel("Total Matches Played by Player")
#y-axis label
plt.ylabel("Maximum run in Player Carier")

plt.show()
```

<h2><center>Display Multiple Plots</center></h2>

<p>With the <code>subplot()</code> function you can draw multiple plots in one figure:</p>

```python
#plot 1:
x = np.array([0, 10, 20, 30])
y = np.array([30, 80, 10, 100])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

<code>plt.subplot(1, 2, 1)</code>
<p>the figure has 1 row, 2 columns, and this plot is the first plot.</p>

<code>plt.subplot(1, 2, 2)</code>
<p>the figure has 1 row, 2 columns, and this plot is the second plot.</p>

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```

<h2><center>Super Title</center></h2>

<h2><center>Matplotlib Scatter</center></h2>

<p>With <code>Pyplot</code>, you can use the <code>scatter()</code> function to draw a scatter plot.

The <code>scatter()</code> function plots one dot for each observation. It needs two arrays of the same length, one for the values of the <code>x-axis</code>, and one for values on the <code>y-axis</code>:</p>

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

<h2><center>Compare Plots</center></h2>

<p>
In the example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well? Will the scatter plot tell us something else?</p>

```python

# Total T20 Matches Played by a Player, and Maximum Runs in Match
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#Total One Day Matches Played by a Player, and Maximum Runs in Match
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.title("Player Performance in Test and One Day Match")
#x-axis label
plt.xlabel("Total Matches Played by Player")
#y-axis label
plt.ylabel("Maximum run in Player Carier")
plt.show()
```

<h2><center>Matplotlib Bars</center></h2>

<p>With Pyplot, you can use the <code>bar()</code> function to draw bar graphs:</p>

```python
x = np.array(["R. Sharma", "V. kohli", "S. Bharat", "A. Patel"])
y = np.array([120, 80, 100, 110])
plt.title("Players Runs in One Match")
plt.bar(x,y)
plt.show()
```

<h2><center>Matplotlib Pie Charts</center></h2>

<p>With Pyplot, you can use the <code>pie()</code> function to draw pie charts:</p>

```python
y = np.array([120, 80, 100, 110])
mylabels = ["R. Sharma", "V. kohli", "S. Bharat", "A. Patel"]

plt.pie(y, labels = mylabels)
plt.title("Top 4 Contributors in Winning Match with Contribution")
plt.legend()
plt.show() 
```
