<h2><center>Distplots</center></h2>

<p>Distplot stands for distribution plot, it takes as input an array and plots a curve corresponding to the distribution of points in the array.</p>

<p>If you want to ignore unnecessary warnings run following commands<br><code>import warnings
warnings.filterwarnings('ignore')</code></p>

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show()
```

<h5>Drawing Displot without Histogram</h5>

```python
sns.distplot([10, 15, 21, 34, 49, 60], hist=False)

plt.show()
```

<h5>Scatterplot with Regression Line</h5>
<p>A scatter plot with a regression line can be used to visualize the relationship between two variables.</p>

```python
import pandas as pd

df = pd.read_csv('data.csv')

# df.head() if a built in function in Python, which will by default show the first 5 Rows
df.head()
```

```python
# Plot between Duration and Pulse

# create a scatterplot with regression line
sns.lmplot(x="Duration", y="Pulse", data=df)

# show the plot
plt.show()
```

```python
# Plot between Maxpulse and Pulse

# create a scatterplot with regression line
sns.lmplot(x="Maxpulse", y="Pulse", data=df)

# show the plot
plt.show()
```

```python
#Box-Plot  b/w MaxPulse and Pulse

sns.boxplot(x="Maxpulse", y="Pulse", data=df)

# show the plot
plt.show()
```

```python
#Box-Plot  b/w Duration and Pulse

sns.boxplot(x="Duration", y="Pulse", data=df)

# show the plot
plt.show()
```

```python
# create a violinplot
sns.violinplot(x="Duration", y="Pulse", data=df)

# show the plot
plt.show()
```

```python
# create a violinplot
sns.violinplot(x="Duration", y="Calories", data=df)

# show the plot
plt.show()
```
