<h2><code>info()</code></h2>

<p>The <code>info()</code> method is a function in Pandas, a popular data manipulation library in Python. It is used to obtain a concise summary of a Pandas DataFrame, including its size, data types, and memory usage. When called on a DataFrame, the <code>info()</code> method provides the number of rows and columns in the dataset, the name and data type of each column, and the amount of memory used by the DataFrame. It is particularly useful for identifying the data types of each column, as well as any missing values or data inconsistencies. The <code>info()</code> method is an essential tool for data exploration and data cleaning, as it provides a quick overview of the structure and content of a dataset.</p>

```python
df.info()
```

<h2><code>describe()</code></h2>

<p>The <code>describe()</code> method is a useful function in Pandas, a popular data manipulation library in Python. It is used to quickly summarize the distribution and statistics of a dataset. When called on a Pandas DataFrame, the <code>describe()</code> method returns a summary of the central tendency, dispersion, and shape of the dataset. It includes information such as the count, mean, standard deviation, minimum and maximum values, and quartiles for each numerical column in the DataFrame. The <code>describe()</code> method is particularly useful for gaining a quick understanding of the structure and content of a dataset, and for identifying any outliers or missing values.</p>

```python
df.describe()
```
